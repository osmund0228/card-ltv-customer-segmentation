```python
import pandas as pd
df = pd.read_csv('df_final.csv')
df.head()
```

# MERCHANT_STATE 기준 분석

```python
state_to_region = {
    # 동북부 (Northeast)
    'CT': 'Northeast', 'ME': 'Northeast', 'MA': 'Northeast', 'NH': 'Northeast', 'RI': 'Northeast', 'VT': 'Northeast', 'NJ': 'Northeast', 'NY': 'Northeast', 'PA': 'Northeast',
    # 중서부 (Midwest)
    'IL': 'Midwest', 'IN': 'Midwest', 'MI': 'Midwest', 'OH': 'Midwest', 'WI': 'Midwest', 'IA': 'Midwest', 'KS': 'Midwest', 'MN': 'Midwest', 'MO': 'Midwest', 'NE': 'Midwest', 'ND': 'Midwest', 'SD': 'Midwest',
    # 남부 (South)
    'DE': 'South', 'FL': 'South', 'GA': 'South', 'MD': 'South', 'NC': 'South', 'SC': 'South', 'VA': 'South', 'DC': 'South', 'WV': 'South', 'AL': 'South', 'KY': 'South', 'MS': 'South', 'TN': 'South', 'AR': 'South', 'LA': 'South', 'OK': 'South', 'TX': 'South',
    # 서부 (West)
    'AZ': 'West', 'CO': 'West', 'ID': 'West', 'MT': 'West', 'NV': 'West', 'NM': 'West', 'UT': 'West', 'WY': 'West', 'AK': 'West', 'CA': 'West', 'HI': 'West', 'OR': 'West', 'WA': 'West'
}

# 가맹점 위치(MERCHANT_STATE) 기준으로 REGION 칼럼 추가
df['REGION'] = df['MERCHANT_STATE'].map(state_to_region).fillna('Unknown')
```

```python
import pandas as pd

# 데이터프레임 이름이 'df'라고 가정합니다.
# 결측치나 공백 제거 등 기본 전처리가 되어있다고 가정합니다.

# ==========================================
# 1. 지역별 업종(카테고리) 소비 비중 비교
# ==========================================
# 각 지역 내에서 특정 카테고리가 차지하는 비중(%)을 계산합니다.
category_ratio = pd.crosstab(
    index=df['MERCHANT_CATEGORY_LEVEL_3'], 
    columns=df['REGION'], 
    normalize='columns' # 열(지역) 기준으로 100%가 되도록 정규화
) * 100

# 보기 편하게 소수점 둘째 자리까지 반올림
category_ratio = category_ratio.round(2)

print("--- 지역별 업종 비중(%) ---")
display(category_ratio)


# ==========================================
# 2. 지역별 가맹점 이름 소비 비중 비교 (전체)
# ==========================================
merchant_ratio = pd.crosstab(
    index=df['MERCHANT_NAME'], 
    columns=df['REGION'], 
    normalize='columns'
) * 100

merchant_ratio = merchant_ratio.round(2)


# ==========================================
# 3. [추천] 각 지역별 TOP 10 가맹점 비중 추출
# ==========================================
# 전체 가맹점을 보면 표가 너무 길어지므로, 지역별로 가장 많이 소비한 TOP 10만 뽑아냅니다.
def get_top_merchants(group, top_n=10):
    # 그룹(지역) 내 비중 계산
    ratios = group.value_counts(normalize=True) * 100
    # 상위 N개 반환
    return ratios.nlargest(top_n).round(2)

# REGION으로 그룹화한 뒤 MERCHANT_NAME 단위로 함수 적용
top_merchants_by_region = df.groupby('REGION')['MERCHANT_NAME'].apply(get_top_merchants)

# 결과를 데이터프레임 형태로 깔끔하게 변환
top_merchants_df = top_merchants_by_region.reset_index()
top_merchants_df.columns = ['REGION', 'MERCHANT_NAME', 'RATIO(%)']

print("\n--- 지역별 TOP 10 가맹점 비중(%) ---")
display(top_merchants_df)
```

```python
import pandas as pd

# df는 현재 사용 중인 데이터프레임 이름이라고 가정합니다.

# 1. 군집별 전체 카드 타입 결제 비율(%) 계산
card_ratio_by_cluster = pd.crosstab(
    index=df['CLUSTER_K4'], 
    columns=df['CARD_TYPE'], 
    normalize='index' # 각 군집의 총 결제 건수를 100%로 두고 정규화
) * 100

# 소수점 둘째 자리까지 반올림
card_ratio_by_cluster = card_ratio_by_cluster.round(2)

print("--- 군집별 전체 카드 타입 비율(%) ---")
display(card_ratio_by_cluster)

# 2. 신용카드(Credit) 비율만 추출해서 보기
# 만약 데이터의 값이 대소문자가 다르다면(예: 'credit') 그에 맞게 수정해 주세요.
if 'Credit' in card_ratio_by_cluster.columns:
    credit_ratio = card_ratio_by_cluster[['Credit']]
    print("\n--- 군집별 신용카드(Credit) 집중 분석(%) ---")
    display(credit_ratio)
else:
    print("\n'Credit' 컬럼이 존재하지 않습니다. CARD_TYPE의 실제 고유값들을 확인해 주세요.")
```
